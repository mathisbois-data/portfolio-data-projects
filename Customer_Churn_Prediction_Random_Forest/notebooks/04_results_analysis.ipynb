## Summary of the Process

- Data cleaning and preprocessing (handling missing values, encoding categorical variables).
- Feature selection and engineering.
- Model training using Random Forest Classifier.
- Model evaluation based on accuracy, recall, confusion matrix, and feature importance analysis.

## Model Design

The Random Forest algorithm was selected due to its robustness, interpretability, and high performance on structured tabular datasets.

Key reasons for this choice include:

- **Handling mixed data types**: Random Forests natively handle both categorical and numerical features without complex preprocessing.
- **Resistance to overfitting**: By averaging multiple decision trees, Random Forests tend to reduce overfitting compared to single Decision Trees.
- **Feature importance analysis**: The model provides a direct measure of the importance of each feature, enabling better interpretation and business insights.
- **Good performance without heavy hyperparameter tuning**: Even with default parameters, Random Forests offer strong baseline performance, making them an excellent choice for initial modeling.

Alternative models such as Gradient Boosting or Logistic Regression could also be explored to potentially enhance performance.

## Model Performance

- **Accuracy**: `0.78`
- **Recall**: `0.48`

Recall is particularly important for churn prediction because identifying customers at risk of leaving is critical, even if it results in a few false positives. A higher recall means the model is effective at capturing churners, which is preferable for retention strategies.

## Confusion Matrix Interpretation

- **True Positives (TP)**: `181` customers correctly predicted as churners.
- **True Negatives (TN)**:  `921` customers correctly predicted as loyal customers.
- **False Positives (FP)**: `112` customers predicted to churn but who actually stay.
- **False Negatives (FN)**: `193` customers who churned but were not predicted.

## Top 5 Most Important Features

| Feature | Importance |
|:--|:--|
| TotalCharges | 18.4% |
| MonthlyCharges | 17.2% |
| tenure | 15.9% |
| Contract | 8.3% |
| PaymentMethod | 5.0% |

- **TotalCharges (18.4%)**:  
  Customers with higher total lifetime charges are less likely to churn, indicating that long-term customers represent a more stable and loyal base.

- **MonthlyCharges (17.2%)**:  
  Higher monthly charges tend to increase the likelihood of churn. Customers paying more per month are more sensitive to perceived value and are more prone to switch providers if they feel overcharged.

- **Tenure (15.9%)**:  
  Tenure, representing the length of time a customer has been with the service, is inversely related to churn: newer customers are significantly more likely to churn compared to long-term customers.

- **Contract (8.3%)**:  
  Customers with month-to-month contracts are at higher risk of churn compared to those with one- or two-year contracts, highlighting the stabilizing effect of long-term commitments.

- **PaymentMethod (5.0%)**:  
  The method of payment also influences churn: customers using electronic checks show a higher tendency to churn compared to those using automatic bank transfers or credit cards.

## Strategic Business Recommendations

Focusing retention efforts on new customers, high monthly charge segments, and month-to-month contracts could significantly reduce churn. Offering contract incentives and pricing adjustments may be particularly effective. These insights can help the business design targeted retention strategies in order to significantly reduce customer churn and improve overall customer lifetime value.
